```python
import time
import random
import numpy as np
import matplotlib.pylab as plt

tottime = 0.0
t0=time.time()
tot_freq=np.zeros(100)
for i in range(0,100):
    a=np.random.random(1000000)
    
    t0=time.time()
    hist=plt.hist(a,bins=100)
    tot_freq+=hist[0]
    t1=time.time()
    
    tottime += t1-t0
    #print tottime
    del a
    del hist

print tot_freq
print 'This code took %f seconds to run' % (tottime)
    
```
