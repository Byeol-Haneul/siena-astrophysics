```python
h = 6.626e-34
c = 3.0e+8
k = 1.38e-23
```

```python
def planck(wav, T=5000):
    a = 2.0*h*c**2
    b = h*c/(wav*k*T)
    intensity = a/ ( (wav**5) * (np.exp(b) - 1.0) )
    return intensity
```

```python
import scipy.integrate as integrate

result = integrate.quad(lambda wav: planck(wav, T=20000), 0, 1e6)
print(result)
result1 = integrate.quad(lambda wav: planck(wav, T=2000), 0, 1e6)
print(result1)
```
