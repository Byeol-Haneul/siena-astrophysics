### Initial imports

```python
import numpy as np # numerical python
import matplotlib.pyplot as plt # plotting package
%matplotlib inline
```

### Make some fake data

```python
xx = np.random.normal(5.0, 1.0, 1000)
print(xx)
```

```python
_ = plt.hist(xx, bins=50)
```

### Fit the data!

```python
xfit = np.arange(-10.0, 20.0, 0.01)
# also check out np.linspace()
mn = 6.0
sig = 3.0
amp = 80
yfit = amp*np.exp(-0.5*(xfit-mn)**2/sig**2)
```

```python
plt.hist(xx, bins=30)
plt.plot(xfit, yfit)
```
